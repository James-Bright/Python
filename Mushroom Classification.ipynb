```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('mushrooms.csv')
display(df.head())

df.isna().sum() ## no nulls

from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df = df.apply(LabelEncoder().fit_transform)

df.head()
```

```python
X = df.drop(['class'], axis = 1)
Y = df['class']

from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.metrics import confusion_matrix
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 42)
```

```python
dec_tree = tree.DecisionTreeClassifier(random_state = 0, max_depth = 5 )
dec_tree.fit(X_train, Y_train)

print(dec_tree.score(X_test,Y_test))
pred_dt = dec_tree.predict(X_test)

dec_tree_confusion = confusion_matrix(pred_dt, Y_test)
print(dec_tree_confusion)
```

```python
from sklearn.ensemble import RandomForestClassifier
rdn_forest = RandomForestClassifier(max_depth = 5)
rdn_forest.fit(X_train, Y_train)

print(rdn_forest.score(X_test,Y_test))
pred_rf = rdn_forest.predict(X_test)

rdn_forest_confusion = confusion_matrix(pred_rf, Y_test)
print(rdn_forest_confusion)
```
